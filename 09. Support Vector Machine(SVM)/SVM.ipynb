Support Vector Machine (SVM)
The SVM draws a hyperplane in n dimensonal space such that it maximizes margin between classification groups.
Hyperplane tries that the margin between the closets points of different should be as maximum as possible.

Train Using Support Vector Machine (SVM)

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop(['target','flower_name'], axis='columns')
X.head()
```

```python
y = df.target
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
from sklearn.svm import SVC
model = SVC()
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
model.predict([[4.8,3.0,1.5,0.3]])
```

Tune perameters

1. Regularization (C)

```python
model_C = SVC(C=10)
model_C.fit(X_train, y_train)
model_C.score(X_test, y_test)
```

2. Gamma

```python
model_g = SVC(gamma=10)
model_g.fit(X_train, y_train)
model_g.score(X_test, y_test)
```

3. Kernel

```python
model_linear_kernal = SVC(kernel='linear')
model_linear_kernal.fit(X_train, y_train)
```

```python
model_linear_kernal.score(X_test, y_test)
```
